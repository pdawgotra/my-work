```python
columns_name=['user_id','item_id','rating','timestamp']
df=pd.read_csv('u.data',sep="\t",names=columns_name)
```

```python
df
```

```python
movies_title=pd.read_csv('u.item',sep="\|",header=None)
```

```python
movies_title
```

```python
movies_title=movies_title[[0,1]]
movies_title.columns=['item_id','title']
```

```python
movies_title
```

```python
df
```

```python
df=pd.merge(df,movies_title,on="item_id")
```

```python
df
```

```python
ratings=pd.DataFrame(df.groupby('title').mean()['rating'])
```

```python
ratings
```

```python
ratings['num']=pd.DataFrame(df.groupby('title').count()['rating'])
```

```python
ratings
```

```python
movie_mat=df.pivot_table(index="user_id",columns="title",values="rating")
```

```python
movie_mat
```

```python
starwars_user_ratings=movie_mat['Star Wars (1977)']
```

```python
starwars_user_ratings
```

```python
similar_to_starwars=movie_mat.corrwith(starwars_user_ratings)
```

```python
similar_to_starwars
```

```python
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['Correlation'])
```

```python
corr_starwars
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars
```

```python
corr_starwars.sort_values('Correlation', ascending=False)
```

```python
corr_starwars
```

```python
corr_starwars=corr_starwars.join(ratings['num'])
```

```python
corr_starwars
```

```python
corr_starwars[corr_starwars['num']>100].sort_values('Correlation',ascending=False)
```

```python
def predict_movies(movie_name):
    movie_user_ratings=movie_mat[movie_name]
    similar_to_movie=movie_mat.corrwith(movie_user_ratings)
    corr_movie=pd.DataFrame(similar_to_movie,columns=['Correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie=corr_movie.join(ratings['num'])
    prediction=corr_movie[corr_movie['num']>100].sort_values('Correlation',ascending=False)
    return prediction
```

```python
predict_movies('Titanic (1997)')
```

```python
predict_movies('Raiders of the Lost Ark (1981)')
```
